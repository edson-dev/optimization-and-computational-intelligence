```python
import networkx as nx
print(nx.__version__)
import matplotlib.pyplot as plt
```

```python
g = nx.DiGraph()

g.add_nodes_from(['lung', 'tub', 'smoke', 'bronc', 'either', 'xray', 'dysp'])
g.add_edges_from([('lung', 'smoke'), ('lung', 'either'), ('tub', 'either'), ('smoke', 'bronc'), ('bronc', 'dysp'), ('either', 'xray'), ('either', 'dysp')])

nx.draw(g, node_color=["red" if node == 'lung' else "blue" for node in g], with_labels=True, width=2,node_size=1200)
#plt.show()
plt.savefig('result/example.png',bbox_inches='tight')
```
